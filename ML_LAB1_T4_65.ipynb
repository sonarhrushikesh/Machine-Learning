```python
# import the libraries
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns
```

```python
# Read data from CSV file
df=pd.read_csv('mcu_box_office.csv') 
df.head() 
```

```python
# Dimension of dataset in the forms of rows and columns
print(df.shape)  
```

```python
# List the columns
df.columns  
```

```python
# print the columns or Attribute with dtype in dataset 
print(df.info())
```

```python
# description of dataset
df.describe() 
df.describe(include='object')
```

```python
# Count occurrences of each movie title and give the unique counts
df['movie_title'].value_counts()
```

```python
# Count occurrences of each movie title and normalize the counts
df['movie_title'].value_counts(normalize=True)
```

```python
# Display the first few rows of the 'mcu_phase' column
df['mcu_phase'].head() 
```

```python
# Using loc[] to select rows and columns by label
df.loc[0:11,'movie_title':'mcu_phase']  
```

```python
# Using iloc[] to select rows and columns by integer position
df.iloc[0:8,6:10]
```

```python
# Filtered selection
df[df['mcu_phase']== df[df['movie_title'] =='Iron Man']['mcu_phase'].max()]['worldwide_box_office'] 
```

```python
# Sorting DataFrame based on 'worldwide_box_office'
df.sort_values(by='worldwide_box_office').head()
df.sort_values(by=['worldwide_box_office','mcu_phase'],ascending=[True,False]).head()
```

```python
# Mapping dictionary
d={'Iron Man': 0, 'Iron Man 2': 1} 

# Applying mapping to 'movie_title' column
df['movie_title'] = df['movie_title'].map(d) 
df.head()
```

```python
# Mapping dictionary
o={0: 'Iron Man', 1: 'Iron Man 2'} 

# Replacing values in 'movie_title' column
df= df.replace({'movie_title':o}) 
df.head()
```

```python
# Grouping by 'mcu_phase' and calculating summary statistics for 'worldwide_box_office'
df.groupby(by='mcu_phase')['worldwide_box_office'].describe()
```

```python
# Generating the contingency table with normalized values
pd.crosstab(df['movie_title'],df['worldwide_box_office'], normalize=True) 
```

```python
# Creating a pivot table with mean aggregation for 'worldwide_box_office'
df.pivot_table(['movie_title', 'mcu_phase'],['worldwide_box_office'], aggfunc='mean') 
```

## Data Visualisation

### 1. Line Graph

```python
# Line Graph
sns.lineplot(x='mcu_phase', y='worldwide_box_office', data=df)
```

### 2. Bar Graph

```python
# Bar Graph
sns.catplot(x='mcu_phase', data=df, kind='count')
```

### 3. Histogram

```python
# Histogram for 'mcu_phase'

plt.hist(df['mcu_phase'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('mcu_phase')
plt.ylabel('Total Movies')
plt.title('Histogram of mcu_phase')

# Show the plot
plt.show()
```

### 4. Scatter plot

```python
# Create the Scatter plot
df.plot.scatter(x='mcu_phase', y='worldwide_box_office', figsize=(10, 10))
plt.xlabel('mcu_phase')
plt.ylabel('Worldwide Box Office')
plt.title('Worldwide Box Office vs mcu_phase')

# Show the plot
plt.show()
```

### 5. Pie chart

```python
# Pie chart

# Select a categorical variable for the pie chart
category_counts = df['mcu_phase'].value_counts()   

# Create the pie chart
plt.figure(figsize=(8, 6))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Movies Across MCU Phases')
plt.axis('equal')

# Show the plot
plt.show()
```
